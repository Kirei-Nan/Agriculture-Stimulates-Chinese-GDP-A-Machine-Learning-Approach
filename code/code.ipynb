```r
df3<-read.csv("OutputData3.csv",header=TRUE)
```

```r
## Creating index variable 

# Simplify Data
data=cbind(df3[,1:3],df3$GDP)
colnames(data)[colnames(data)=="df3$GDP"] <- "GDP"
```

```r
# Save data for future use
write.csv(data, file = "C:/MyData.csv")
```

```r
# Read data from saved file
data = read.csv("C:/MyData.csv", header=T)
data =data[,-1]
```

```r
##########################################################################################generate train data and test data
```

```r
# # Random sampling
# samplesize = 0.60 * nrow(data)
# set.seed(80)
# index = sample( seq_len ( nrow ( data ) ), size = samplesize )
```

```r
# # Create training and test set
# datatrain = data[ index, ]
# datatest = data[ -index, ]
```

```r
# Create training and test set
datatrain = data %>% slice(1:35)
datatest = data %>% slice(36:39)
```

```r
## Scale data for neural network
## Normalized data to the same scale

max = apply(data , 2 , max)
min = apply(data, 2 , min)
scaled = as.data.frame(scale(data, center = min, scale = max - min))
```

```r
# # creating training and test set
# trainNN = scaled[index , ]
# testNN = scaled[-index , ]
```

```r
trainNN=scaled %>% slice(1:35)
testNN=scaled %>% slice(36:39)
```

```r
###################################################################################### train neural network
```

```r
# load library
library(neuralnet)
```

```r
# fit neural network
set.seed(2)
NN = neuralnet(formula =GDP~Output.of.Persimmons.10000.tons.+Output.of.Apples.10000.tons.+Output.of.Garden.Fruits.10000.tons.,trainNN,hidden=c(4,4,4),linear.output = T )
```

```r
#plot the model
library(NeuralNetTools)
plotnet(NN)
```

```r
plot(NN,rep='best')
```

```r
##########################################################predictTest
```

```r
dataMax=max(data$GDP)
dataMin=min(data$GDP)
```

```r
predict_testNN = compute(NN, testNN[,c(1:3)])
```

```r
predict_testNN = predict_testNN$net.result * (dataMax - dataMin) + dataMin
```

```r
plot(datatest$GDP, col='black')
par(new=TRUE)
points(predict_testNN,col='red')
```

```r
############################################################evaluate
```

```r
preds <- predict_testNN
actual <- datatest$GDP
```

```r
# Calculate Root Mean Square Error (RMSE)
RMSE.NN = (sum((actual - preds)^2) / nrow(datatest)) ^ 0.5
paste("RMSE: ", round(RMSE.NN, 4))
```

```r
Test_SSE <- sum((preds- actual)^2)/2
paste("SSE: ", round(Test_SSE, 4))
```

```r
rss <- sum((preds - actual) ^ 2)  ## residual sum of squares
tss <- sum((actual - mean(actual)) ^ 2)  ## total sum of squares
rsq <- 1 - rss/tss
paste("R-squared: ", round(rsq, 4))
```
